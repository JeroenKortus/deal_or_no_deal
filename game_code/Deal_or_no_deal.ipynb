```python
import random
import time
from IPython.display import clear_output

def bankers_formula(total, roundnr):
    round_offer = [0.30, 0.30, 0.30, 0.30, 0.30, 0.40, 0.50, 0.60, 0.70]
    #offer = 12275.30 + (0.748 * E) + (-2714.74 * C) + (-0.040 * M) + (.0000006986 * (E**2)) + (32.623 * (C**2))
    #if offer < 0.01:
    #    offer = 0.01
    print()
    return total*round_offer[roundnr-1]

def inputNumber(case_number):
    while True:
        try:
            userInput = int(input("Choose a case!\n"))
        except ValueError:
            print("Not an integer! Try again.")
            continue
        else:
            
            if (1 <= userInput <= 26) and (userInput is not case_number): # this is faster
                return userInput
                break
            else:
                print("Out of range or your own case. Try again")

def ask_deal_or_no_deal(offer):
    print("The bank has offered you ${:.2f}.".format(offer))
    print("DEAL OR NO DEAL!")
    while True:
        userInput = input("write \"d\" or \"n\"\n")
        if userInput not in ["d", "n"]:
            print("Please check your spelling")
            continue
        else:
            break
    return userInput

class game:
    def __init__(self):
        # Deal or no deal indicator
        self.deal = False
        
        # The cases
        self.case_values = [0.01, 1, 5, 10, 25, 50, 75, 100, 200, 300, 
               400, 500, 750, 1000, 5000, 10000, 25000, 50000, 75000, 
               100000, 200000, 300000, 400000, 500000, 750000, 1000000]
        random.shuffle(self.case_values)
        self.cases = []
        for c in range(26):
            self.cases.append(case(c+1, self.case_values[c]))
        
        # Players case
        self.case_picked = 27
        
        # Rounds managers
        self.round = 0
        self.round_cases_to_open = [6,5,4,3,2,1,1,1,1]

    def pick_case(self):
        self.case_picked = inputNumber(self.case_picked)


    def start_round(self):
        for pick in range(self.round_cases_to_open[self.round]):
            while True:
                chosen_nr = inputNumber(self.case_picked) - 1
                if self.cases[chosen_nr].opened:
                    print("Case has already been opened.")
                    continue
                self.cases[chosen_nr].open()
                break

        self.round += 1

    def print_cases(self):
        print("{: >6}{: >4}{: >4}{: >4}{: >4}{: >4}".format(*["xx" if i.opened else i.index for i in self.cases[0:6] ]))
        print("{: >4}{: >4}{: >4}{: >4}{: >4}{: >4}{: >4}".format(*["xx" if i.opened else i.index for i in self.cases[6:13] ]))
        print("{: >4}{: >4}{: >4}{: >4}{: >4}{: >4}{: >4}".format(*["xx" if i.opened else i.index for i in self.cases[13:20] ]))
        print("{: >6}{: >4}{: >4}{: >4}{: >4}{: >4}".format(*["xx" if i.opened else i.index for i in self.cases[20:26] ]))
        print("-----------------------------------")
        if self.round != 0:
            print("Your case is {}".format(self.case_picked))
        
        sorted_cases = sorted(self.cases, key=lambda x: x.value)
        for i in range(13):
            print("{: >10}{: >10}".format("xxxx" if sorted_cases[i].opened else sorted_cases[i].value, "xxxx" if sorted_cases[i+13].opened else sorted_cases[i+13].value))
        
    def ask_banker(self):
        total = 0
        nr_cases = 0
        max_value = 0
        #print(nr_cases)
        for c in self.cases:
            if not c.opened:
                total += c.value
                nr_cases += 1
                if c.value > max_value:
                    max_value = c.value
        #print(nr_cases)
        offer = bankers_formula(total/nr_cases, self.round)
        answer = ask_deal_or_no_deal(offer)
        if answer == "d":
            self.deal = True
            print("Congratulations, you have just won ${:.2f}!".format(offer))
            print("Your case contained ${}".format(self.cases[self.case_picked - 1].value))
            
    def ask_switch(self):
        for c in self.cases:
            if not c.opened and c.index is not self.case_picked:
                print("Do you want to switch your case with number {}?".format(c.index))
                userInput = input("write (y)es or (n)o. \n")
                if userInput == "y":
                    print("you have won ${}!".format(c.value))
                else:
                    print("you have won ${}!".format(self.cases[self.case_picked - 1].value))
                break
                
        
        
        
class case:
    def __init__(self, index, value):
        self.index = index
        self.value = value
        self.opened = False
        
    def open(self):
        self.opened = True
        print("The case you chose contained ${}.".format(self.value))
        time.sleep(2)
        
```

```python
game1 = game()
game1.print_cases()
game1.pick_case()

while not game1.deal:
    if game1.round == 9:
        game1.ask_switch()
        break
        
    game1.start_round()
    clear_output(wait=False)
    game1.print_cases()
    game1.ask_banker()
    
```
